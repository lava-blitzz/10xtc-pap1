```python
import pyttsx3

engine=pyttsx3.init()
responses={
    "hello":"Hi there!",
    "how are you":"I'm a computer program, but I'm here to help!",
    "who are you":"I'm a bot",
    "data detective":"The datatypes:",
    "bye":"Goodbye, have a great day or night!"
}

def get_bot_response(user_input):
    says=responses.get(user_input,"I'm not sure how to respond to that.")
    return says

def data_detective(user_input):
    data_type=df.dtypes
    print("Data types of each column:")
    print(data_type)
    return data_type

rate=engine.getProperty('rate')
engine.setProperty('rate',rate-15)
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)
def start_chat():
    print("Bot: Hi there! How can I help you?")
    while True:
        user_input=input("You: ").lower()
        if user_input=="bye":
            print(responses["bye"])
            engine.say("Goodbye, have a great day or night!")
            engine.runAndWait()
            break
        bot_says=get_bot_response(user_input)
        print("Bot:",bot_says)
        engine.say(bot_says)
        engine.runAndWait()
start_chat()
```
